```python
#Project target-data cleaning and finding correlations between gross and other variables of the movies

#Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Read data
df = pd.read_csv('movies1.csv')

#Data Visualization
df
```

```python
#Looping through the data to search for anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Columns data types

print(df.dtypes)
```

```python
#Removing duplicates from the data

df.drop_duplicates()
```

```python
#Order data by gross value

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Searching for outliers

df.boxplot(column=['gross'])
```

```python
#Regression plot with gross vs budget using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#Regression plot with score vs gross using seaborn

sns.regplot(x="score", y="gross", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
#Corellation matrix heatmap visualisation for numeric features

correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#Random numeric value assignment for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
#Corellation matrix heatmap visualisation for movies

correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
