```python
#Data Exploration for Automobile Data set
#Explore features and charecteristics to predict price of car

#Import Libraries

import pandas as pd
import numpy as np
```

```python
#Load data and store in dataframe df


path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

```python
#Import visualisation packages

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#List the data types for each column

print(df.dtypes)
```

```python
#Calculation of correlation between variable

df.corr()

#We can notice that the diagonal elements are always one
```

```python
#Correlation between bore,stroke,compression-ratio and horsepower columns

df[['bore', 'stroke', 'compression-ratio', 'horsepower']].corr()
```

```python
#Investigating the relationship between an individual variable and the price

#Use "regplot", which plots the scatterplot plus the fitted regression line for the data

#Positive linear relationship

#scatterplot of "engine-size" and "price"

sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

```python
#As the engine-size goes up, the price goes up: this indicates a positive direct correlation between these two variables. 
#Engine size is a good predictor of price since the regression line is almost a perfect diagonal line.
```

```python
#Examine the correlation between 'engine-size' and 'price'

df[["engine-size", "price"]].corr()
```

```python
#It's approximately 0.87
```

```python
#Highway mpg is a potential predictor variable of price

sns.regplot(x="highway-mpg", y="price", data=df)
```

```python
#As the highway-mpg goes up, the price goes down: this indicates an inverse/negative relationship between these two variables.
#Highway mpg could potentially be a predictor of price.
```

```python
#Correlation between 'highway-mpg' and 'price

df[['highway-mpg', 'price']].corr()
```

```python
#It's approximately -0.704 
```

```python
#Weak Linear Relationship

#"Peak-rpm" as a predictor variable of "price"

sns.regplot(x="peak-rpm", y="price", data=df)
```

```python
#Peak rpm does not seem like a good predictor of the price at all since the regression line is close to horizontal. 
#Also, the data points are very scattered and far from the fitted line, showing lots of variability. 
#Therefore it's it is not a reliable variable.
```

```python
#Correlation between 'peak-rpm' and 'price' 

df[['peak-rpm','price']].corr()
```

```python
#Relationship between "body-style" and "price"

sns.boxplot(x="body-style", y="price", data=df)
```

```python
#We see that the distributions of price between the different body-style categories have a significant overlap, 
#and so body-style would not be a good predictor of price. 
```

```python
#Relationship between"engine-location" and "price"

sns.boxplot(x="engine-location", y="price", data=df)
```

```python
#Here we see that the distribution of price between these two engine-location categories, front and rear, 
#are distinct enough to take engine-location as a potential good predictor of price.
```

```python
#Relationship between "drive-wheels" and "price"

sns.boxplot(x="drive-wheels", y="price", data=df)
```

```python
#Here we see that the distribution of price between the different drive-wheels categories differs
#as such drive-wheels could potentially be a predictor of price.
```

```python
#Convert series to a dataframe

df['drive-wheels'].value_counts().to_frame()
```

```python
#Save the results to the dataframe "drive_wheels_counts" and rename the column 'drive-wheels' to 'value_counts'

drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.rename(columns={'drive-wheels': 'value_counts'}, inplace=True)
drive_wheels_counts
```

```python
#Rename the index to 'drive-wheels'

drive_wheels_counts.index.name = 'drive-wheels'
drive_wheels_counts
```

```python
#Group by the variable "drive-wheels"

df['drive-wheels'].unique()


#There are 3 different categories of drive wheels
```

```python
#Finding type of drive wheel which is most valuable by average

df_group_one = df[['drive-wheels','body-style','price']]
```

```python
#Grouping results

df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
#It seems rear-wheel drive vehicles are, on average, the most expensive, 
#while 4-wheel and front-wheel are approximately the same in price
```

```python
#Grouping results
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
#This grouped data is much easier to visualize when it is made into a pivot table

grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
#Average "price" of each car based on "body-style"

df_gptest2 = df[['body-style','price']]
grouped_test_bodystyle = df_gptest2.groupby(['body-style'],as_index= False).mean()
grouped_test_bodystyle
```

```python
#Visualize how the price is related to 'drive-wheel' and 'body-style'

plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
